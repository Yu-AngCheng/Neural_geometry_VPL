```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pickle
from matplotlib import ticker
from scipy.stats import sem, ttest_rel
from RZutilpy.figure import bar, colorpalettable, default_img_set
default_img_set()
```

```python
ryzcolor = colorpalettable('ryz1')
C0, C1, C0_alpha, C1_alpha = np.array(ryzcolor(0)), np.array(ryzcolor(1)),np.array(ryzcolor(0)), np.array(ryzcolor(1))
C0_alpha[-1] = 0.3
C1_alpha[-1] = 0.3
Clist = [np.array(ryzcolor(2)), np.array(ryzcolor(3)), np.array([0.121, 0.6039, 0.227]), np.array(ryzcolor(1))]
Calphalist = [np.array(ryzcolor(2)), np.array(ryzcolor(3)), np.array([0.121, 0.6039, 0.227, 1.0]), np.array(ryzcolor(1))]
Calphalist[0][-1]=0.3
Calphalist[1][-1]=0.3
Calphalist[2][-1]=0.3
Calphalist[3][-1]=0.3
with open('../Results/Metrics_All_Pre_Post.pkl', 'rb') as f:
    Metrics = pickle.load(f)
PCA_LFI = Metrics["PCA_LFI"].mean(axis=(1,5))[:,:,0,0,:]
PCA_LFI_mean = PCA_LFI.mean(axis=(0))
PCA_LFI_se = sem(PCA_LFI,axis=0)
fig, ax = plt.subplots(1, 5,figsize=(15, 4))
for i in range(5):
    bh=bar(1, PCA_LFI_mean[i:i+1,:], yerr=PCA_LFI_se[i:i+1,:], bargap=0.03, ecolor='k', axes=ax[i], color=Calphalist, edgecolor=Clist, linewidth=2)
    ax[i].set_xticks([])
    ax[i].set_xticklabels([])
    # ax[i].set_ylim([-0.6, 3.0])
    ax[i].spines['bottom'].set_position(('data', 0))
ax[0].set_ylabel('Information gain')
ax[0].set_ylim([-0.6, 1.5])
ax[1].set_ylim([-0.6, 2.5])
ax[2].set_ylim([-0.6, 3.0])
ax[3].set_ylim([-0.3, 3.5])
ax[4].set_ylim([0.0, 3.5])
plt.savefig('FigS6_1.pdf')
```
